```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('housing_area_price.csv')
df


plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')

new_df = df.drop('price',axis='columns')
new_df

price = df.price
price


reg = linear_model.LinearRegression()
reg.fit(new_df,price)
reg.predict([[3300]])
reg.coef_
reg.intercept_
reg.predict([[5000]])
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('canada_per_capita_income.csv')

# Rename columns for easy access
df.columns = ["year", "per_capita_income"]

# Visualizing the data
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df.year, df.per_capita_income, color='red', marker='+')
plt.show()

# Prepare data for training
X = df[['year']]  # Feature
y = df['per_capita_income']  # Target variable

# Train linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict per capita income for the year 2020
predicted_income_2020 = model.predict([[2020]])
print(f"Predicted per capita income for 2020: ${predicted_income_2020[0]:,.2f}")

# Get model parameters
print(f"Coefficient (m): {model.coef_[0]}")
print(f"Intercept (b): {model.intercept_}")

# Verify using Y = mX + b
manual_calculation = 2020 * model.coef_[0] + model.intercept_
print(f"Manual Calculation: ${manual_calculation:,.2f}")
```
