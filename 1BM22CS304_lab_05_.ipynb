```python
# ðŸŽ¯ KNN Implementation for IRIS, DIABETES, and HEART datasets ðŸš€

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler

print("\nðŸŸ¢ " + "KNN on IRIS Dataset".upper())
# Load the iris dataset (ensure "iris.csv" is in your working directory)
iris_df = pd.read_csv("iris.csv")
print("IRIS Columns:", iris_df.columns.tolist())

# Define target column: 'species' if it exists, otherwise use the last column
target_col_iris = 'species' if 'species' in iris_df.columns else iris_df.columns[-1]
X_iris = iris_df.drop(columns=[target_col_iris])
y_iris = iris_df[target_col_iris]

# Split the data (80% train, 20% test)
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.2, random_state=42)

# Choose k (here, 5) and train the model
k_iris = 5
knn_iris = KNeighborsClassifier(n_neighbors=k_iris)
knn_iris.fit(X_train_iris, y_train_iris)

# Predict & Evaluate
y_pred_iris = knn_iris.predict(X_test_iris)
accuracy_iris = accuracy_score(y_test_iris, y_pred_iris)
cm_iris = confusion_matrix(y_test_iris, y_pred_iris)
print(f"Accuracy Score (IRIS): {accuracy_iris:.4f}")
print("Confusion Matrix (IRIS):")
print(cm_iris)
print("Classification Report (IRIS):")
print(classification_report(y_test_iris, y_pred_iris))

# Plot Confusion Matrix for IRIS
plt.figure(figsize=(6,5))
sns.heatmap(cm_iris, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix for IRIS Dataset (KNN)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Plot Classification Report for IRIS
cr_iris_dict = classification_report(y_test_iris, y_pred_iris, output_dict=True)
cr_df_iris = pd.DataFrame(cr_iris_dict).transpose()
if 'support' in cr_df_iris.columns:
    cr_df_iris = cr_df_iris.drop(columns=['support'])
plt.figure(figsize=(8,5))
sns.heatmap(cr_df_iris, annot=True, cmap="Greens", fmt=".2f")
plt.title("Classification Report Heatmap for IRIS Dataset (KNN)")
plt.show()




print("\nðŸŸ¢ " + "KNN on DIABETES Dataset".upper())
# Load the diabetes dataset (ensure "diabetes.csv" is available)
diabetes_df = pd.read_csv("diabetes.csv")
print("DIABETES Columns:", diabetes_df.columns.tolist())

# Define target column: 'Outcome' if exists, otherwise use the last column
target_col_dia = 'Outcome' if 'Outcome' in diabetes_df.columns else diabetes_df.columns[-1]
X_diabetes = diabetes_df.drop(columns=[target_col_dia])
y_diabetes = diabetes_df[target_col_dia]

# Feature Scaling (crucial for KNN)
scaler = StandardScaler()
X_diabetes_scaled = scaler.fit_transform(X_diabetes)

# Split the data
X_train_dia, X_test_dia, y_train_dia, y_test_dia = train_test_split(X_diabetes_scaled, y_diabetes, test_size=0.2, random_state=42)

# Train KNN with k = 5
k_dia = 5
knn_dia = KNeighborsClassifier(n_neighbors=k_dia)
knn_dia.fit(X_train_dia, y_train_dia)

# Predict & Evaluate
y_pred_dia = knn_dia.predict(X_test_dia)
accuracy_dia = accuracy_score(y_test_dia, y_pred_dia)
cm_dia = confusion_matrix(y_test_dia, y_pred_dia)
print(f"Accuracy Score (DIABETES): {accuracy_dia:.4f}")
print("Confusion Matrix (DIABETES):")
print(cm_dia)
print("Classification Report (DIABETES):")
print(classification_report(y_test_dia, y_pred_dia))

# Plot Confusion Matrix for DIABETES
plt.figure(figsize=(6,5))
sns.heatmap(cm_dia, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix for DIABETES Dataset (KNN)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Plot Classification Report for DIABETES
cr_dia_dict = classification_report(y_test_dia, y_pred_dia, output_dict=True)
cr_df_dia = pd.DataFrame(cr_dia_dict).transpose()
if 'support' in cr_df_dia.columns:
    cr_df_dia = cr_df_dia.drop(columns=['support'])
plt.figure(figsize=(8,5))
sns.heatmap(cr_df_dia, annot=True, cmap="Greens", fmt=".2f")
plt.title("Classification Report Heatmap for DIABETES Dataset (KNN)")
plt.show()



print("\nðŸŸ¢ " + "KNN on HEART Dataset".upper())
# Load the heart dataset (ensure "heart.csv" is available)
heart_df = pd.read_csv("heart.csv")
print("HEART Columns:", heart_df.columns.tolist())

# Define target column: try 'target' or 'Target', otherwise use the last column
if 'target' in heart_df.columns:
    target_col_heart = 'target'
elif 'Target' in heart_df.columns:
    target_col_heart = 'Target'
else:
    target_col_heart = heart_df.columns[-1]
X_heart = heart_df.drop(columns=[target_col_heart])
y_heart = heart_df[target_col_heart]

# Feature Scaling for KNN
scaler_heart = StandardScaler()
X_heart_scaled = scaler_heart.fit_transform(X_heart)

# Split the data
X_train_heart, X_test_heart, y_train_heart, y_test_heart = train_test_split(X_heart_scaled, y_heart, test_size=0.2, random_state=42)

# Hyperparameter Tuning: Find the best k value (from 1 to 20)
k_values = range(1, 21)
best_accuracy = 0
best_k = 0
accuracy_list = []
for k in k_values:
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train_heart, y_train_heart)
    y_pred_temp = knn_temp.predict(X_test_heart)
    acc = accuracy_score(y_test_heart, y_pred_temp)
    accuracy_list.append(acc)
    if acc > best_accuracy:
        best_accuracy = acc
        best_k = k

print(f"Best k for HEART dataset: {best_k} with accuracy: {best_accuracy:.4f}")

# Train final model with best_k
knn_heart_final = KNeighborsClassifier(n_neighbors=best_k)
knn_heart_final.fit(X_train_heart, y_train_heart)
y_pred_heart = knn_heart_final.predict(X_test_heart)

# Evaluate the model
cm_heart = confusion_matrix(y_test_heart, y_pred_heart)
print(f"Accuracy Score (HEART): {accuracy_score(y_test_heart, y_pred_heart):.4f}")
print("Confusion Matrix (HEART):")
print(cm_heart)
print("Classification Report (HEART):")
print(classification_report(y_test_heart, y_pred_heart))

# Plot Confusion Matrix for HEART
plt.figure(figsize=(6,5))
sns.heatmap(cm_heart, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix for HEART Dataset (KNN)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Plot Classification Report for HEART
cr_heart_dict = classification_report(y_test_heart, y_pred_heart, output_dict=True)
cr_df_heart = pd.DataFrame(cr_heart_dict).transpose()
if 'support' in cr_df_heart.columns:
    cr_df_heart = cr_df_heart.drop(columns=['support'])
plt.figure(figsize=(8,5))
sns.heatmap(cr_df_heart, annot=True, cmap="Greens", fmt=".2f")
plt.title("Classification Report Heatmap for HEART Dataset (KNN)")
plt.show()
```
